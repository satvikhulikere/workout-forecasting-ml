```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.callbacks import EarlyStopping
from sklearn.metrics import mean_squared_error
import os

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("../workout-forecasting-ml/data/workout_data.csv")
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['Exercise', 'Date'])

# Features to include (independent variables)
feature_cols = ['Reps', 'Sets', 'SleepHours', 'ProteinIntake', 'Calories', 'SorenessLevel', 'Bodyweight']

# -------------------------------
# 2. Create multivariate sequences
# -------------------------------
def create_sequences_multivariate(features, target, seq_length):
    x, y = [], []
    for i in range(len(target) - seq_length):
        x.append(features[i:i + seq_length])
        y.append(target[i + seq_length])
    return np.array(x), np.array(y)

# -------------------------------
# 3. Loop through each exercise
# -------------------------------
save_dir = "../workout-forecasting-ml"
os.makedirs(save_dir, exist_ok=True)

for exercise in df["Exercise"].unique():
    print(f"\n---- Training Multivariate LSTM for: {exercise} ----")
    
    ex_df = df[df["Exercise"] == exercise].copy().reset_index(drop=True)

    if len(ex_df) < 20:
        print("Not enough data. Skipping.")
        continue

    # Combine features and target
    full_cols = feature_cols + ['WeightLifted']
    ex_df = ex_df[full_cols + ['Date']].dropna().reset_index(drop=True)

    # Scale features + target together
    scaler = MinMaxScaler()
    scaled = scaler.fit_transform(ex_df[feature_cols + ['WeightLifted']])
    scaled_df = pd.DataFrame(scaled, columns=feature_cols + ['WeightLifted'])

    # Separate features (multivariate) and target
    features_scaled = scaled_df[feature_cols].values
    target_scaled = scaled_df['WeightLifted'].values

    # Create sequences
    sequence_length = 5
    x, y = create_sequences_multivariate(features_scaled, target_scaled, sequence_length)

    # Reshape x to [samples, time steps, features]
    x = x.reshape((x.shape[0], sequence_length, len(feature_cols)))

    # -------------------------------
    # 4. Build LSTM model
    # -------------------------------
    model = Sequential()
    model.add(LSTM(64, input_shape=(sequence_length, len(feature_cols))))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train with early stopping
    es = EarlyStopping(monitor='loss', patience=10, restore_best_weights=True)
    model.fit(x, y, epochs=100, batch_size=8, verbose=1, callbacks=[es])

    # -------------------------------
    # 5. Evaluate model
    # -------------------------------
    preds = model.predict(x)
    
    # Inverse transform predictions and actual
    dummy = np.zeros((len(preds), len(feature_cols) + 1))
    dummy[:, -1] = preds.flatten()
    predicted_weights = scaler.inverse_transform(dummy)[:, -1]

    dummy[:, -1] = y
    actual_weights = scaler.inverse_transform(dummy)[:, -1]

    rmse = np.sqrt(mean_squared_error(actual_weights, predicted_weights))
    print(f"{exercise} RMSE: {rmse:.2f}")

    # Plot actual vs predicted
    plt.figure(figsize=(10, 5))
    plt.plot(ex_df['Date'][sequence_length:], actual_weights, label='Actual', marker='o')
    plt.plot(ex_df['Date'][sequence_length:], predicted_weights, label='Predicted', linestyle='--', marker='x')
    plt.title(f"{exercise} — Actual vs Predicted")
    plt.xlabel("Date")
    plt.ylabel("Weight Lifted")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # -------------------------------
    # 6. Save the trained model
    # -------------------------------
    model_path = os.path.join(save_dir, f"{exercise}.h5")
    model.save(model_path)
    print(f"✅ Saved model: {model_path}")
```
