```python
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
from keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
import os

# Constants
DATA_PATH = "/Users/satvikhulikere/workout-forecasting-ml/data/workout_data.csv"
MODEL_DIR = "/Users/satvikhulikere/workout-forecasting-ml/models"
SEQUENCE_LENGTH = 5

# Features used during training
FEATURE_COLS = ['Reps', 'Sets', 'SleepHours', 'ProteinIntake', 'Calories', 'SorenessLevel', 'Bodyweight']

# Load workout data
@st.cache_data
def load_data():
    df = pd.read_csv(DATA_PATH)
    df["Date"] = pd.to_datetime(df["Date"])
    return df

# Load trained models
@st.cache_resource
def load_models():
    models = {}
    for file in os.listdir(MODEL_DIR):
        if file.endswith(".h5"):
            exercise = file.replace(".h5", "")
            models[exercise] = load_model(os.path.join(MODEL_DIR, file))
    return models

# Main app logic
def main():
    st.title("🏋️‍♂️ Lift Forecasting with LSTM")
    st.markdown("Visualize and forecast your lifts using personalized LSTM models.")

    df = load_data()
    models = load_models()

    exercises = list(models.keys())
    selected_exercise = st.sidebar.selectbox("Choose an exercise:", exercises)

    df_ex = df[df["Exercise"] == selected_exercise].sort_values("Date").reset_index(drop=True)

    if len(df_ex) < SEQUENCE_LENGTH + 1:
        st.warning("Not enough data for this exercise.")
        return

    # Normalize features + target
    scaler = MinMaxScaler()
    scaled = scaler.fit_transform(df_ex[FEATURE_COLS + ["WeightLifted"]])
    df_scaled = pd.DataFrame(scaled, columns=FEATURE_COLS + ["WeightLifted"])

    # Create sequences
    def create_sequences(data, seq_len):
        x, y = [], []
        for i in range(len(data) - seq_len):
            x.append(data[i:i+seq_len, :-1])
            y.append(data[i+seq_len, -1])
        return np.array(x), np.array(y)

    data = df_scaled[FEATURE_COLS + ["WeightLifted"]].values
    x, y = create_sequences(data, SEQUENCE_LENGTH)

    model = models[selected_exercise]
    preds = model.predict(x)
    actual = y.reshape(-1, 1)

    # Inverse transform only WeightLifted
    dummy = np.zeros((len(actual), len(FEATURE_COLS) + 1))
    dummy[:, -1] = actual[:, 0]
    actual_inv = scaler.inverse_transform(dummy)[:, -1]

    dummy[:, -1] = preds[:, 0]
    preds_inv = scaler.inverse_transform(dummy)[:, -1]

    # Plot actual vs predicted
    st.subheader(f"{selected_exercise} — Actual vs Predicted")
    fig1, ax1 = plt.subplots(figsize=(10, 4))
    ax1.plot(df_ex["Date"][SEQUENCE_LENGTH:], actual_inv, label="Actual", marker='o')
    ax1.plot(df_ex["Date"][SEQUENCE_LENGTH:], preds_inv, label="Predicted", linestyle='--', marker='x')
    ax1.set_ylabel("Weight Lifted")
    ax1.set_xlabel("Date")
    ax1.set_title(f"{selected_exercise} — Actual vs Predicted")
    ax1.legend()
    st.pyplot(fig1)

    # -------------------------------
    # Forecast next 90 days
    # -------------------------------
    st.subheader("📈 Forecast for Next 90 Days")
    forecast_preds = []
    future_dates = [df_ex["Date"].iloc[-1] + timedelta(days=i+1) for i in range(90)]

    # Use last sequence for forecasting
    last_seq = df_scaled.iloc[-SEQUENCE_LENGTH:].values
    current_seq = last_seq[:, :-1]  # exclude WeightLifted
    current_seq = current_seq.reshape(1, SEQUENCE_LENGTH, len(FEATURE_COLS))

    for _ in range(90):
        next_pred_scaled = model.predict(current_seq)[0][0]
        forecast_preds.append(next_pred_scaled)

        # Build next input sequence
        next_features = current_seq[0, -1, :].copy()
        dummy_input = np.append(next_features, next_pred_scaled)

        # Append predicted row and reshape sequence
        full_next_row = dummy_input[:-1]  # drop predicted weight from input features
        next_seq = np.concatenate((current_seq[0, 1:, :], full_next_row.reshape(1, -1)), axis=0)
        current_seq = next_seq.reshape(1, SEQUENCE_LENGTH, len(FEATURE_COLS))

    # Inverse transform forecasted weights
    dummy_forecast = np.zeros((90, len(FEATURE_COLS) + 1))
    dummy_forecast[:, -1] = forecast_preds
    forecast_inv = scaler.inverse_transform(dummy_forecast)[:, -1]

    forecast_df = pd.DataFrame({
        "Date": future_dates,
        "Predicted Weight Lifted": forecast_inv
    })

    # Plot forecast
    fig2, ax2 = plt.subplots(figsize=(10, 4))
    ax2.plot(forecast_df["Date"], forecast_df["Predicted Weight Lifted"], marker='o', label="Forecast")
    ax2.set_title(f"{selected_exercise} — 90-Day Forecast")
    ax2.set_ylabel("Predicted Weight Lifted")
    ax2.set_xlabel("Date")
    ax2.legend()
    st.pyplot(fig2)

    # Show forecast table
    st.dataframe(forecast_df.set_index("Date"))

if __name__ == "__main__":
    main()
```
